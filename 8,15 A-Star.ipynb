```python
def goalState(initial, goal):
    for i in range(0, len(initial)):
        for j in range(0, len(goal)):
            if initial[i][j] != goal[i][j]:
                return False
    return True
```

```python
def check(matrix):
    for i in range(len(matrix)):
        for j in range(len(matrix[i])):
            if matrix[i][j] == 0:
                return i, j
```

```python
def swap_elements(matrix, r1, c1, r2, c2):
    matrix[r1][c1], matrix[r2][c2] = matrix[r2][c2], matrix[r1][c1]
```

```python
def generate_child(state):
    children = []
    r, c = check(state)
    possible_moves = [(r - 1, c), (r + 1, c), (r, c - 1), (r, c + 1)]
    for move in possible_moves:
        new_r, new_c = move
        if 0 <= new_r < R and 0 <= new_c < C:
            child = [row.copy() for row in state]
            swap_elements(child, r, c, new_r, new_c)
            children.append(child)
    return children
```

```python
def herustic(curr, final):
  count = 0
  for i in range(len(curr)):
    for j in range(len(final)):
      if curr[i][j] != final[i][j]:
        count = count + 1
  return count
```

```python
def herustic_man(curr, final):
  dist = 0
  for i in range(len(curr)):
    for j in range(len(final)):
      val = curr[i][j]

  return dist
```

```python
def astar(initial, final):
    open_list = [(0 + herustic(initial , final), 0, initial)]
    closed_list = []


    while open_list:
        open_list.sort(key = lambda x : x[0])
        f, cost, curr_state = open_list.pop(0)

        print("Current state:")
        for row in curr_state:
            print(row)

        if goalState(curr_state, final):
            print("Reached goal state\n")
            return cost

        if curr_state in closed_list:
            continue

        closed_list.append(curr_state)
        for child in generate_child(curr_state):
            if child not in closed_list and child not in open_list:
                open_list.append((1 + herustic(child , final), cost + 1, child))

    print("Goal state not reached.")
```

```python
def count_inversions(matrix):
    inv_count = 0
    a = [element for row in matrix for element in row]

    for i in range(len(a)):
        for j in range(i + 1, len(a)):
            if a[i] != 0 and a[j] != 0 and a[i] > a[j]:
                inv_count += 1

    return inv_count
```
